```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data_sorted[['child_mort','exports','health','imports',
             'income','inflation','total_fer', 'gdpp']] = scaler.fit_transform(data_sorted[['child_mort','exports','health',
                                                                                            'imports','income','inflation','total_fer','gdpp']])
```

```python
data_sorted
```

```python
from sklearn.cluster import KMeans

kmeans_pred = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(data_sorted)
```

```python
kmeans_pred
```

```python
kmeans_pred.labels_
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters = k)
    km.fit(data_sorted)
    sse.append(km.inertia_)
```

```python
sse
```

```python
from kneed import KneeLocator

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
sns.lineplot(x=k_range, y=sse, color='#355070')

plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.xticks(k_range)
plt.title("Elbow Method to find best K")

kl = KneeLocator(k_range, sse, curve='convex', direction='decreasing')
plt.axvline(x=kl.elbow, color='#40916c', linestyle='--', label='Elbow Point')

plt.legend()
plt.show()
```

```python
kmeans_pred_elbow = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(data_sorted)
```

```python
kmeans_pred_elbow.labels_
```

```python
data_sorted
```

```python
data_sorted['cluster'] = kmeans_pred_elbow.labels_
```

```python
cluster_centers = kmeans_pred_elbow.cluster_centers_
cluster_centers
```

```python
def check_result(data,plot=False):
    if plot:
        sns.set_style('darkgrid')
        fig, ax = plt.subplots(1, 2)
        ax = np.reshape(ax, (1, 2))
        ax[0, 0] = sns.countplot(data=data, x='cluster', palette="Blues", ax=ax[0, 0])
        ax[0, 0].set_ylabel('Count',fontsize=14)
        ax[0, 0].set_xticklabels(ax[0, 0].get_xticklabels(), rotation=0)
        slices = data['cluster'].value_counts().sort_index().values
        activities = [var for var in data['cluster'].value_counts().sort_index().index]
        ax[0, 1] = plt.pie(slices, labels=activities,
                           colors=sns.color_palette('Blues'), autopct='%1.01f%%')
        plt.title("Percentage",fontsize=14)
        fig.set_size_inches(12, 6)
        fig.suptitle('Count of Each Cluster', fontsize=14)
        plt.show()
        
check_result(data_sorted, plot=True)
```

```python
import scipy.cluster.hierarchy as sch

fig=plt.figure(figsize=(15, 7))
sch.dendrogram(sch.linkage(data_sorted, method='ward',metric='euclidean'))
plt.title('Dendrogram')
plt.xlabel('Rows')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
df1 = data_sorted[data_sorted.cluster == 0]
df2 = data_sorted[data_sorted.cluster == 1]
df3 = data_sorted[data_sorted.cluster == 2]

for idx, column in enumerate(data_sorted.columns):
    if column == 'gdpp':
        break
    else:
        plt.scatter(df1['gdpp'], df1[column], color='green', label='Cluster 0') 
        plt.scatter(df2['gdpp'], df2[column], color='red', label='Cluster 1') 
        plt.scatter(df3['gdpp'], df3[column], color='blue', label='Cluster 2')
        plt.scatter(cluster_centers[:, -1], cluster_centers[:, idx], color='black', marker='*', label='Centroids') 
        plt.xlabel('gdpp')
        plt.ylabel(column)
        plt.legend()
        plt.title(f'Cluster Scatter Plot for {column}')
        plt.show()
```
